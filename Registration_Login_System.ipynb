```python
import csv
import pandas as pd
# Continue the loop as long as any one of the action (Login or regsitration) is completed fully.
success = 0
while success == 0:
    # Get user selection.
    vuinp = 0
    while vuinp == 0:
        uinp = input("Enter any one option. Either Login or Register: ")
        if uinp != "Login" and uinp != "Register":
            print("Enter a Valid Option.")
        else:
            vuinp = 1
    print("User selected:", uinp)
    # if user selection is Register.
    if uinp == "Register":
        vUI = 0
        vPW = 0
        # get UI & check whether it matches the criteria
        while vUI == 0:
            UI = input("Enter the Mail ID to be registered: ")
            if UI.find("@")==-1 or UI.find(".")== -1:
                print("Incorrect Mail ID. Please check the Mail ID.")
            elif UI.find('@') <= 1:
                print("Incorrect Mail ID. Please check the Mail ID.")
            elif (UI.find('.') -  UI.find('@')) == 1:
                print("Incorrect Mail ID. Please check the Mail ID.")
            elif UI[0].isalpha() == False:
                print("Incorrect Mail ID. Please check the Mail ID.")
            else:
                print("Mail ID matches the expected criteria.")
                vUI = 1
        # get PW & check whether it matches the criteria        
        while vPW == 0:
            PW = input("Enter the PW:")
            isN = 0
            isL = 0
            isU = 0
            isS = 0
            for i in range(0,len(PW)):
                if PW[i].isdigit() == True:
                    isN = 1
                elif PW[i].isupper() == True:
                    isU = 1
                elif PW[i].islower() == True:
                    isL = 1
                elif PW[i].isalpha() == False:
                    isS = 1
            if isN ==0 or isL == 0 or isU == 0 or isS == 0:
                print("PW does not match the expected criteria.")
            elif len(PW) > 16 or len(PW)< 5:
                print("PW length should be minimum of 6 Characters and maximum of 16 characters.")
            else:
                print("PW matches the required criteria.")
                vPW = 1
        # Checked whether the UI already exists in the file. 
        # If yes throw an error message.
        # if not then add to the file.
        from csv import writer
        rows = [UI, PW]
        pr = 0
        with open('Register.csv', mode ='r')as file:
            csvFile = csv.reader(file)
            for lines in csvFile:
                if UI in lines:
                    pr = 1
                    break
        if pr == 1:
            print("UI already registered. Please try registering using some other ID. Or try logging in.")
        else:
            with open('Register.csv','a',newline='') as file:
                csvwriter = writer(file)
                csvwriter.writerow(rows)
                file.close()
            print("User Registered successfully!!!")
            success = 1
    
    # if user selection is Login.
    elif uinp == "Login":
        # get the UI & PW from the user
        UI = input("Enter the Mail ID to log in: ")
        PW = input("Enter the PW:")
        UIpr = 0
        PWpr = 0
        # check whether the UI & PW exists in the file.
        # If not, then ask the user to register.
        # if yes then accept the user crendtials.
        with open('Register.csv', mode ='r')as file:
            csvFile = csv.reader(file)
            for lines in csvFile:
                if UI in lines:
                    UIpr = 1
                    if PW in lines:
                        PWpr=1
                        break
        if UIpr==0:
            print("Incorrect UI. Please re-check the UI. If not registered, please register.")
        elif PWpr==0:
            while PWpr == 0:
                print("Incorrect PW. Please re-set your PW.")
                PW = input("Enter the PW:")
                df=pd.read_csv("Register.csv")
                df.loc[df['UI'] == UI, 'PW'] = PW
                df.to_csv("Register.csv", index=False)
                print("PW successfully reset!!!")
                PWpr = 1
                success = 1
        else:
            print("Login Successful!!!")
            success = 1
            
print("Thanks for using the application!!!")
```
